```python
import pandas as pd
from pathlib import Path
```

```python
# Get the current notebook directory
notebook_dir = Path.cwd()  # This points to the folder where the notebook is running

# Construct the correct path to your CSV
data_path = notebook_dir / "../data/raw/sales_records.csv"

# Resolve full absolute path (optional, just for debugging)
data_path = data_path.resolve()
print("Reading from:", data_path)

# Load dataset
df = pd.read_csv(data_path)

# Show first 5 rows
df.head()
```

```python
#data types of each column as expected, but if we want to do date-based analysis, we need to change the date columns to datetime format
df.dtypes 
```

```python
#only needed to change the date columns to datetime format for future analysis
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df.dtypes
```

```python
#Checking for Order Date range
print("Order Date range:", df["Order Date"].min(), "to", df["Order Date"].max())
print("Ship Date range:", df["Ship Date"].min(), "to", df["Ship Date"].max())
```

```python
#Checking missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
#CHecking for duplicates
print("Number of duplicate rows:", df.duplicated().sum())  
```
