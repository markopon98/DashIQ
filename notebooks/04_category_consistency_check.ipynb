```python
import pandas as pd
from pathlib import Path
```

```python
notebook_dir = Path.cwd()  

data_path = notebook_dir / "../data/raw/sales_records.csv"

data_path = data_path.resolve()
print("Reading from:", data_path)

df = pd.read_csv(data_path)

df.head()
```

```python
# View unique Item Type values
df["Item Type"].unique()    
```

```python
# Count category frequencies
df["Item Type"].value_counts()
```

```python
# Compare raw vs cleaned category values
raw_categories = df["Item Type"].unique()

cleaned_categories = (
    df["Item Type"]
    .astype(str)
    .str.strip()
    .str.title()
    .unique()
)

print("Raw unique category count:", len(raw_categories))
print("Cleaned unique category count:", len(cleaned_categories))
```

```python
# View unique region values
df["Region"].unique()
```

```python
# View unique region values
df["Region"].unique()
```

```python
# Compare raw vs cleaned region values
raw_regions = df["Region"].unique()

cleaned_regions = (
    df["Region"]
    .astype(str)
    .str.strip()
    .str.title()
    .unique()
)

print("Raw unique region count:", len(raw_regions))
print("Cleaned unique region count:", len(cleaned_regions))
```

```python
# confirming columns
df.columns.tolist()
```

```python
# Check for category values with leading/trailing spaces
category_whitespace_issues = df[df["Item Type"] != df["Item Type"].str.strip()]
category_whitespace_issues["Item Type"].unique()
```

```python
# Check for region values with leading/trailing spaces
region_whitespace_issues = df[df["Region"] != df["Region"].str.strip()]
region_whitespace_issues["Region"].unique()
```

```python
# Create standardized preview columns
df["Item Type Clean"] = df["Item Type"].astype(str).str.strip().str.title()
df["Region Clean"] = df["Region"].astype(str).str.strip().str.title()

df[["Item Type", "Item Type Clean", "Region", "Region Clean"]].head()
```
