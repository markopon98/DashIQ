```python
import pandas as pd
from pathlib import Path
```

```python
# Load clean Data
notebook_dir = Path.cwd()  
data_path = notebook_dir / "../data/cleaned/sales_records_clean.csv"

data_path = data_path.resolve()
print("Reading from:", data_path)

df = pd.read_csv(data_path)


df["Order Date"] = pd.to_datetime(df["Order Date"])

df.head()
```

```python
df["Year"] = df["Order Date"].dt.year
```

```python
df["Month"] = df["Order Date"].dt.month
```

```python
df["Month"] = df["Order Date"].dt.month
```

```python
# Avoiding division error
df["Profit Margin"] = df["Total Profit"] / df["Total Revenue"]
df["Profit Margin"] = df["Profit Margin"].fillna(0)
```

```python
df["Profit Margin (%)"] = df["Profit Margin"] * 100
```

```python
def discount_bucket(margin):
    if margin < 0.1:
        return "High Discount"
    elif margin < 0.3:
        return "Medium Discount"
    else:
        return "Low Discount"

df["Discount Bucket"] = df["Profit Margin"].apply(discount_bucket)
```
